Keep in mind, this is not a true robust recommendation system, to describe it more accurately,it just tells you what movies/items are most similar to your movie choice.

### Recommending Similar Movies

```python
df.head()
```

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
```

```python
moviemat
```

```python
ratings.sort_values("num of ratings", ascending=False)
```

**Now, we suppose a people watch the Star Wars, then we want to recommend the most similiar movie for him**

```python
starwar_user_ratings = moviemat["Star Wars (1977)"]
```

```python
starwar_user_ratings
```

We can then use corrwith() method to get correlations between two pandas series:

```python
similar_to_starwar = moviemat.corrwith(starwar_user_ratings,axis=0)
```

```python
similar_to_starwar
```

While the sample variance is equal to 0, results will return NaN

```python
corr_starwars = pd.DataFrame(similar_to_starwar,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars
```

除了按照相关性高低的顺序排列，我们还需要考虑剔除那些没人看或者评分很低的电影

```python
corr_starwars = pd.merge(corr_starwars,ratings,on="title")
corr_starwars.head()
```

```python
recommender = corr_starwars[(corr_starwars["rating"]>3) & (corr_starwars["num of ratings"]>100)]
```

```python
recommender.sort_values("Correlation",ascending=False)
```

**Therefore, we will suggest the Empire Strikes Back, The (1980) for this people**
