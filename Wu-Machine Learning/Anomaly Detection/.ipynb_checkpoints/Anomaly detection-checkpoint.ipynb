### Estimating parameters for Gaussian distribution

```python
#可以选择是否使用多元高斯分布
def getGaussianParams(X,useMultivariate):
    mu = X.mean(axis=0)#按列求均值
    if useMultivariate:
        sigma2 = ((X-mu).T @ (X-mu))/len(X)#X-mu python会自动广播 ##计算协方差矩阵
    else:
        sigma2 = X.var(axis=0,ddof=0)#ddof=0 对应分母为m ddof=1对应分母为m-1
    return mu, sigma2
```

### Calculate Gaussian density function value

```python
def Gaussian(X, mu, sigma2):
    '''
    mu, sigma2参数已经决定了一个高斯分布模型
    因为原始模型就是多元高斯模型在sigma2上是对角矩阵而已，所以如下：
    If Sigma2 is a matrix, it is treated as the covariance matrix. 
    If Sigma2 is a vector, it is treated as the sigma^2 values of the variances
    in each dimension (a diagonal covariance matrix)
    output:
        一个(m, )维向量，包含每个样本的概率值。
    '''
    m,n = X.shape
    #np.ndim()求维度，当是一维的时候，转化为对角矩阵
    if np.ndim(sigma2) == 1:
        sigma2 = np.diag(sigma2)
    constant = 1/(np.power((2*np.pi),n/2)*np.sqrt(np.linalg.det(sigma2)))
    '''
    当矩阵过大时，numpy矩阵相乘会出现内存错误，这里我们逐行进行运算 
    '''
    exp = np.zeros((m,1))
    for i in range(m):
        Xrow = X[i]
        exp[i] = np.exp(-0.5*((Xrow-mu).T @ np.linalg.inv(sigma2) @ (Xrow-mu)))
    return constant * exp
```

### 画高斯概率分布图

```python
def plotContours(mu,sigma2):
    delta = 0.1   
    x = np.arange(0,30,delta)
    y = np.arange(0,30,delta)
    
    xx,yy = np.meshgrid(x,y)
    points = np.c_[xx.ravel(),yy.ravel()]#按列合并
    z = Gaussian(points,mu,sigma2)
    z = z.reshape(xx.shape)
    cont_levels = [10**h for h in range(-20,0,3)]
    plt.contour(xx,yy,z,cont_levels)
```

```python
# First contours without using multivariate gaussian:
plotData()
useMV = False
# *表示解元组
plotContours(*getGaussianParams(X, useMV))

# Then contours with multivariate gaussian:
plotData()
useMV = True
plotContours(*getGaussianParams(X, useMV))
```

### Select the threshold ε

在这部分的练习中,您将实现一个算法使用交叉验证集的F1分数来选择合理的阈值 ε 

```python
def selectThreshold(yval,Pval):
    bestEpsilon, bestF1 = 0,0
    stepsize = (max(Pval) - min(Pval)) / 1000
    for epsilon in np.arange(min(Pval),max(Pval),stepsize):
        pred_bool = Pval < epsilon
        predictions = pred_bool.astype("int").ravel()
        fp = sum((predictions == 1) & (yval == 0))
        fn = sum((predictions == 0) & (yval == 1))
        tp = sum((predictions == 1) & (yval == 1))
        prec = tp/(tp+fp) if (tp+fp) else 0
        rec = tp/(tp+fn) if (tp+fp) else 0
        F1 = 2*prec*rec/(prec+rec) if (prec+rec) else 0
        if F1 > bestF1:
            bestF1 = F1
            bestEpsilon = epsilon
    return bestF1, bestEpsilon
```

```python
mu, sigma2 = getGaussianParams(X,useMultivariate=False)
Pval = Gaussian(Xval,mu,sigma2)
bestF1, bestEpsilon = selectThreshold(yval,Pval)
```

```python
bestF1,bestEpsilon
```

```python
pred_bool = Pval < 0.1
predictions = pred_bool.astype("int").ravel()
fp = sum((predictions == 1) & (yval == 0))
fn = sum((predictions == 0) & (yval == 1))
tp = sum((predictions == 1) & (yval == 1))
```

Then we apply the threshold we find to original X

```python
mu, sigma2 = getGaussianParams(X,useMultivariate=False)
Probability = Gaussian(X,mu,sigma2)
xx = np.array([X[i] for i in range(len(Probability)) if Probability[i] < bestEpsilon])
```
