### 变量初始化

```python
X = np.insert(X,0,1,axis=1) #在X的index为0的地方按列（axis=1）插入一列1
X.shape
y = y.ravel() #refine y的格式，使其适用于算法
y.shape
```

### Idea

我们最后的目标是得到一个5000x10的矩阵，行对应着每一组数据，列对应着10个不同的logistic regression而算出来的P的值，而我们要去找最大的那个P的值，将数据分为对应的类。<br/>
X在加上X0=1后的维度是5000x401，所以我们要找到一个维度是401x10的theta。<br/>
为了寻找theta，我们可以做10次logistic regression，然后将每一次算出的theta向量放在一个矩阵的一列中。

### Sigmoid function

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

### Regularized Cost function

```python
def regularized_cost(theta,X,y,lamda):
    first = (-y) * np.log(sigmoid(np.dot(X,theta)))#   * 表示array对应元素相乘
    second = (1 - y)*np.log(1 - sigmoid(np.dot(X,theta)))
    part1 = np.mean(first - second)
    part2 = (lamda/(2*len(X)))*(theta[1:]@theta[1:])#一维array内积
    return part1+part2
```

### Regularized gradient function

```python
def regularized_gradient(theta,X,y,lamda):
    grad = (X.T @ (sigmoid(np.dot(X,theta))-y))/len(X)#   @ 表示array的矩阵乘法和np.dot()是一样的
    reg = (lamda/len(X))*theta
    reg[0] = 0
    return grad + reg
```

### One-vs-all classification

```python
 import scipy.optimize as opt
```

```python
def one_vs_all(X,y,lamda,K):
    final_theta = np.zeros((X.shape[1],K)) #final_theat (401,10)
    for i in range(1,K+1):
        theta = np.zeros(X.shape[1]) #theta (401,)
        yi = np.array([1 if value == i else 0 for value in y]) #yi (5000,)
        result = opt.fmin_tnc(func=regularized_cost,x0=theta,fprime=regularized_gradient, args=(X, yi, lamda))
        final_theta[:,i-1]=result[0]
    return final_theta
```

```python
final_theta = one_vs_all(X,y,2,10)
```

### Prediction accuarcy

```python
def predict(X,final_theta):
    h = np.dot(X,final_theta)
    index = np.argmax(h,axis=1)#返回每一行最大值的下标
    prediction = index+1
    return prediction
```

```python
y_predict = predict(X,final_theta)
accuracy = np.mean(y_predict==y)
print("accuracy= {:.2f}%".format(accuracy*100))#保留两位小数
```
