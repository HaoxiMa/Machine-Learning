# Neural networks (Weights are offered)

Data：<br/>
有5000个训练样本，每个样本是20*20像素的数字的灰度图像。每个像素代表一个浮点数，表示该位置的灰度强度。20×20的像素网格被展开成一个400维的向量。在我们的数据矩阵X中，每一个样本都变成了一行，这给了我们一个5000×400矩阵X，每一行都是一个手写数字图像的训练样本。<br/>
同时，我们给出了神经网络的Weights。<br/>
The parameters have dimensions that are sized for a neural network with 25 units in the second layer and 10 output units (corresponding to the 10 digit classes).

### 数据预处理

```python
from scipy.io import loadmat #package for loading data in MATLAB format
```

```python
data = loadmat("data.mat")
X = data["X"]
y = data["y"]
type(X),type(y)
X.shape,y.shape
```

```python
Weights = loadmat("weights.mat")
Theta1 = Weights["Theta1"]
Theta2 = Weights["Theta2"]
```

```python
type(Theta1),type(Theta2)
Theta1.shape,Theta2.shape
```

### Sigmoid function

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

### Neural Networks

```python
X = np.insert(X,0,1,axis=1)
```

```python
X.shape
```

```python
Z2 = np.dot(X,Theta1.T)
```

```python
Z2.shape
```

```python
a2 = sigmoid(Z2)
a2.shape
```

```python
a2 = np.insert(a2,0,1,axis=1)
a2.shape
```

```python
Z3 = np.dot(a2,Theta2.T)
```

```python
a3 = sigmoid(Z3)
a3.shape
```

```python
y_predict = np.argmax(a3,axis=1)+1  #y_predict 为array
accuracy = np.mean(y_predict==y.flatten()) #flatten()函数抹平y
print("accuracy = {:.2f}%".format(accuracy*100))
```

机器学习速度快准确率很高，但可能存在overfit
